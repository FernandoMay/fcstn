## ═══ FREQUENCY → MANDELBROT MAPPING ═══

Each cognitive state vector is projected onto 5 frequency bands, which are then weighted and summed to produce a unique `(cx, cy, zoom)` coordinate in the Mandelbrot set.

| Band | Freq Range | Cognitive Correlate | Mandelbrot Mapping |
|------|-----------|-------------------|-------------------|
| δ | 0.5–4 Hz | Deep sleep, unfocused | `cx: -1.5, zoom: 0.5` |
| θ | 4–8 Hz | Relaxation, meditation | `cx: -1.0, cy: 0.3` |
| α | 8–13 Hz | Focus, calm attention | `cx: -0.5, cy: 0.0, zoom: 2.0` |
| β | 13–30 Hz | Active engagement | `cx: -0.75, cy: 0.15, zoom: 4.0` |
| γ | 30–50 Hz | Complex cognition | `cx: -0.16, cy: 1.04, zoom: 6.0` |

```python
# === CELL: Cognitive State → Frequencies → Mandelbrot ===
cog_states = {
    'RESTING':  {'attention': 0.3, 'engagement': 0.2, 'workload': 0.1, 'valence': 0.5, 'coherence': 0.4},
    'FOCUSED':  {'attention': 0.9, 'engagement': 0.7, 'workload': 0.4, 'valence': 0.6, 'coherence': 0.8},
    'STRESSED': {'attention': 0.6, 'engagement': 0.8, 'workload': 0.9, 'valence': 0.2, 'coherence': 0.3},
    'CREATIVE': {'attention': 0.7, 'engagement': 0.6, 'workload': 0.5, 'valence': 0.8, 'coherence': 0.5},
    'ENGAGED':  {'attention': 0.8, 'engagement': 0.9, 'workload': 0.6, 'valence': 0.7, 'coherence': 0.7},
}

print(f'{"State":<12} {"δ":>6} {"θ":>6} {"α":>6} {"β":>6} {"γ":>6}  {"cx":>8} {"cy":>8} {"zoom":>6}  {"Label"}')
print('-' * 80)

fig, axes = plt.subplots(1, 2, figsize=(14, 5), facecolor='#0A0E14')
ax1, ax2 = axes
ax1.set_facecolor('#0A0E14')
ax2.set_facecolor('#0A0E14')

band_colors = ['#667', '#88a', '#00DBE7', '#4EDEA3', '#f80']
band_labels = ['\u03b4', '\u03b8', '\u03b1', '\u03b2', '\u03b3']

for i, (sname, state) in enumerate(cog_states.items()):
    freqs = profile.cognitive_to_frequencies(state)
    j = profile.freq_to_journey(freqs)
    print(f'{sname:<12} {freqs["delta"]:>6.2f} {freqs["theta"]:>6.2f} {freqs["alpha"]:>6.2f} {freqs["beta"]:>6.2f} {freqs["gamma"]:>6.2f}  {j["cx"]:>8.4f} {j["cy"]:>8.4f} {j["zoom"]:>6.2f}  {j["label"]}')
    
    # Plot spectrum
    vals = [freqs[b] for b in ['delta','theta','alpha','beta','gamma']]
    ax1.bar([x + i*0.12 for x in range(5)], vals, width=0.1, label=sname if i==0 else '',
            color=[f'hsl({j["hue_shift"] + x*30}, 70%, {40 + i*10}%)' for x in range(5)], alpha=0.8)

ax1.set_xticks(range(5))
ax1.set_xticklabels(band_labels, fontfamily='monospace', color='#B9CACB')
ax1.set_ylabel('Amplitude', color='#B9CACB', fontfamily='monospace')
ax1.set_title('Frequency Spectrum per Cognitive State', color='#00DBE7', fontfamily='monospace', fontsize=11)
ax1.legend(loc='upper right', facecolor='#0A0E14', labelcolor='#B9CACB', fontsize=8)
ax1.set_ylim(0, 1.1)

# Plot 2D Mandelbrot coordinate mapping
for sname, state in cog_states.items():
    freqs = profile.cognitive_to_frequencies(state)
    j = profile.freq_to_journey(freqs)
    ax2.scatter(j['cx'], j['cy'], s=j['zoom']*30, alpha=0.7, 
               label=f'{sname} (z={j["zoom"]:.1f})')
    ax2.annotate(sname, (j['cx'], j['cy']), 
                color='#B9CACB', fontsize=8, fontfamily='monospace',
                xytext=(5, 5), textcoords='offset points')

ax2.set_xlabel('cx', color='#B9CACB', fontfamily='monospace')
ax2.set_ylabel('cy', color='#B9CACB', fontfamily='monospace')
ax2.set_title('Mandelbrot Coordinate Map', color='#00DBE7', fontfamily='monospace', fontsize=11)
ax2.legend(loc='lower right', facecolor='#0A0E14', labelcolor='#B9CACB', fontsize=8)
ax2.grid(True, alpha=0.1)
ax2.set_xlim(-1.6, 0.6)
ax2.set_ylim(-0.4, 1.4)

plt.tight_layout()
plt.show()
```

## ═══ 3D FRACTAL TUNNEL ═══

The tunnel visualization interpolates between consecutive Mandelbrot positions, creating a 3D flight path through the set. Each frame is stacked with perspective (scale + opacity decay) to simulate depth.

```python
# === CELL: Tunnel Animation ===
def generate_tunnel_frames(n_frames=12, width=320, height=240, palette='cyberpunk'):
    """Generate a sequence of Mandelbrot frames for tunnel animation."""
    # Simulate cognitive state evolution over time
    frames = []
    for i in range(n_frames):
        t = i / max(1, n_frames - 1)
        # Oscillate cognitive parameters
        attn = 0.4 + 0.5 * math.sin(t * math.pi * 3)
        eng = 0.3 + 0.6 * math.cos(t * math.pi * 2.5)
        load = 0.2 + 0.3 * abs(math.sin(t * math.pi * 4))
        val = 0.5 + 0.3 * math.sin(t * math.pi * 2)
        coh = 0.4 + 0.4 * math.cos(t * math.pi * 1.5)
        
        state = {'attention': attn, 'engagement': eng, 'workload': load,
                 'valence': val, 'coherence': coh}
        freqs = profile.cognitive_to_frequencies(state)
        j = profile.freq_to_journey(freqs)
        
        img = render_mandelbrot(width, height, palette, 96,
                                j['cx'], j['cy'], max(0.5, j['zoom']),
                                0.0, 0.0)
        frames.append({'img': img, 'cx': j['cx'], 'cy': j['cy'],
                       'zoom': j['zoom'], 'label': j['label']})
    return frames

frames = generate_tunnel_frames(12)
print(f'Generated {len(frames)} tunnel frames')

# Render tunnel composite
W, H = frames[0]['img'].size
tunnel = Image.new('RGB', (W, H), (10, 14, 20))

for i, f in enumerate(frames):
    depth = i / max(1, len(frames) - 1)
    scale = 1.0 - depth * 0.35
    alpha = int((1.0 - depth * 0.55) * 255)
    
    dw = int(W * scale)
    dh = int(H * scale)
    img_resized = f['img'].resize((dw, dh), Image.LANCZOS)
    
    # Add hue shift
    r, g, b = img_resized.split()
    img_resized = Image.merge('RGB', (
        r.point(lambda x: int(x * alpha / 255)),
        g.point(lambda x: int(x * alpha / 255)),
        b.point(lambda x: int(x * alpha / 255)),
    ))
    
    dx = (W - dw) // 2
    dy = (H - dh) // 2
    tunnel.paste(img_resized, (dx, dy), img_resized)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5), facecolor='#0A0E14')
ax1.set_facecolor('#0A0E14')
ax2.set_facecolor('#0A0E14')

ax1.imshow(tunnel)
ax1.set_title('3D TUNNEL COMPOSITE — All Frames Stacked', color='#00DBE7', fontfamily='monospace', fontsize=10)
ax1.axis('off')

# Plot the 2D trajectory
cxs = [f['cx'] for f in frames]
cys = [f['cy'] for f in frames]
zooms = [f['zoom'] for f in frames]
sc = ax2.scatter(cxs, cys, c=range(len(frames)), cmap='plasma', s=[z*15 for z in zooms], alpha=0.8)
ax2.plot(cxs, cys, '--', color='#00DBE7', alpha=0.3, linewidth=0.5)
for i, f in enumerate(frames):
    ax2.annotate(f'{i}', (f['cx'], f['cy']), color='#B9CACB', fontsize=7, fontfamily='monospace',
                xytext=(3, 3), textcoords='offset points')
ax2.set_xlabel('cx', color='#B9CACB', fontfamily='monospace')
ax2.set_ylabel('cy', color='#B9CACB', fontfamily='monospace')
ax2.set_title('TUNNEL TRAJECTORY — Frequency-Driven Path', color='#00DBE7', fontfamily='monospace', fontsize=10)
ax2.grid(True, alpha=0.1)
plt.colorbar(sc, ax=ax2, label='Frame', shrink=0.6)
plt.tight_layout()
plt.show()
```

## ═══ ANIMATED FLIGHT ═══

Watch the Mandelbrot evolve as cognitive state oscillates. Each frame is a new frequency-mapped position.

```python
# === CELL: Animated Flight Through Mandelbrot ===
def animate_flight(n_frames=30, palette='cyberpunk', interval=300):
    fig, ax = plt.subplots(figsize=(10, 7.5), facecolor='#0A0E14')
    ax.set_facecolor('#0A0E14')
    ax.axis('off')
    im = ax.imshow(np.zeros((300, 400, 3), dtype=np.uint8))
    
    info_text = ax.text(0.02, 0.96, '', transform=ax.transAxes,
                       fontfamily='monospace', fontsize=8, color='#00DBE7',
                       verticalalignment='top',
                       bbox=dict(boxstyle='round,pad=0.3', facecolor='#0A0E14', alpha=0.8, edgecolor='#00DBE7'))
    
    title = ax.set_title('', color='#4EDEA3', fontfamily='monospace', fontsize=10)
    
    def update(frame):
        t = frame / max(1, n_frames - 1)
        attn = 0.3 + 0.6 * (0.5 + 0.5 * math.sin(t * math.pi * 4))
        eng = 0.3 + 0.6 * (0.5 + 0.5 * math.cos(t * math.pi * 3))
        load = 0.2 + 0.5 * abs(math.sin(t * math.pi * 5))
        val = 0.5 + 0.3 * math.sin(t * math.pi * 2 + 1.0)
        coh = 0.5 + 0.4 * math.cos(t * math.pi * 2.5)
        
        state = {'attention': attn, 'engagement': eng, 'workload': load,
                 'valence': val, 'coherence': coh}
        freqs = profile.cognitive_to_frequencies(state)
        j = profile.freq_to_journey(freqs)
        
        img = render_mandelbrot(400, 300, palette, 96,
                               j['cx'], j['cy'], max(0.5, j['zoom']), 0.0, 0.0)
        im.set_array(img)
        
        info = f'FRAME {frame+1}/{n_frames}  |  α: {freqs["alpha"]:.2f}  β: {freqs["beta"]:.2f}  γ: {freqs["gamma"]:.2f}'
        info_text.set_text(info)
        title.set_text(f'FCSTN FLIGHT — cx: {j["cx"]:.4f}  cy: {j["cy"]:.4f}  zoom: {j["zoom"]:.2f}  [{j["label"]}]')
        return [im, info_text, title]
    
    ani = animation.FuncAnimation(fig, update, frames=n_frames, interval=interval, blit=True)
    return ani

ani = animate_flight(20, 'cyberpunk', 400)
display(HTML(ani.to_jshtml()))
```
